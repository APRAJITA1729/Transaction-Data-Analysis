### Calculate KPIs

#### Total Numbers of Transactions

```python
transactions_count = df["TransactionID"].count()
print("Total Numbers of  Transactions:", transactions_count)
```

#### Total Transaction amount

```python
total_transactions_amount = df["Amount"].sum()
print("Total Transaction:",total_transactions_amount)
```

#### Average Transaction 

```python
average_transaction = total_transactions_amount/transactions_count
print(average_transaction)
```

#### Highest Transaction amount

```python
highest_transaction = df["Amount"].max()
print(highest_transaction)
```

#### Top Sending Bank (by Amount)

```python
top_sending_bank_amount = df.groupby('BankNameSent')['Amount'].sum().idxmax()
top_sending_amount = df.groupby('BankNameSent')['Amount'].sum().max()

print("Top Sending Bank (by Amount):", top_sending_bank_amount)
print("Total Amount Sent:", top_sending_amount)
```

#### Top Receiving Bank (by Amount)

```python
top_receiving_bank_amount = df.groupby('BankNameReceived')['Amount'].sum().idxmax()
top_receiving_bank_amount= df.groupby('BankNameReceived')['Amount'].sum().max()

print("Top Receiving Bank (by Amount):", top_receiving_bank_amount)
print("Total Amount Received:", top_receiving_bank_amount)
```

#### Monthly Transactions Tend

```python
import calendar

# Ensure Month column exists
df['Month'] = df['TransactionDate'].dt.month

# Group by month number
monthly_value = df.groupby('Month')['Amount'].sum()

# Replace month number with short name (Jan, Feb, ...)
monthly_value.index = monthly_value.index.map(lambda x: calendar.month_abbr[x])

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(monthly_value.index, monthly_value, marker='o', linestyle='-', color='b')

plt.title('Monthly Trascations Trend', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)

plt.show()
```

#### Top merchants by spend

```python
# Group by Merchant and calculate total spend
merchant_spend = df.groupby('MerchantName')['Amount'].sum().sort_values(ascending=False)

# Select Top 10 Merchants
top_merchants = merchant_spend.head(3)

# Plot bar chart
plt.figure(figsize=(10,6))
top_merchants.plot(kind='bar', color='violet', edgecolor='black')

plt.title("Top 3 Merchants by Spend", fontsize=14)
plt.xlabel("Merchant", fontsize=12)
plt.ylabel("Total Spend (â‚¹)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
